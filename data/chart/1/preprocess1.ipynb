```python
import pandas as pd
import numpy as np
```

```python
conf_data = pd.read_csv('../conflict_data_mmr.csv')
conf_data.head()
```

```python
gov_conflict = conf_data[(conf_data['side_a'] == 'Government of Myanmar (Burma)') & (conf_data['side_b'] != 'Civilians') & (conf_data['side_b'] != 'NUG')]

lat = gov_conflict['latitude'].to_numpy()
lon = gov_conflict['longitude'].to_numpy()
sideb = gov_conflict['side_b'].to_numpy()

unique_sideb = np.unique(sideb)
print(f"Unique side_b values: {unique_sideb} {len(unique_sideb)}")

pd.DataFrame({'latitude': lat, 'longitude': lon, 'side_b': sideb}).to_csv('coords_mmr.csv', index=False)
```
