```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../MultiCountry WDD.xlsx")
data.head()
```

```python
rows_to_keep = ['GDP per capita (current US$)',
                'GDP (current US$)',
                'Population, total',
                'Merchandise exports (current US$)',
                'Merchandise imports (current US$)',
                'Government Effectiveness: Percentile Rank',
                'Agriculture, forestry, and fishing, value added (current US$)',
                'Industry (including construction), value added (current US$)',
                'Military expenditure (% of GDP)']

data_filtered = data[data['Series Name'].isin(rows_to_keep)]

#remove nigeria
data_filtered = data_filtered[data_filtered['Country Name'] != 'Nigeria']

#export to csv
data_filtered.to_csv("countries_indicators.csv", index=False)
```
