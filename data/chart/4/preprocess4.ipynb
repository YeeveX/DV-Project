```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_excel("../apac_aggregated.xlsx")
data.head()
```

```python
my_data = data[data['COUNTRY'] == 'Myanmar']

my_data['YEAR'] = my_data['WEEK'].dt.year
my_data['MONTH'] = my_data['WEEK'].dt.month
my_data['DAY'] = my_data['WEEK'].dt.day

after_coup = my_data[(my_data['YEAR'] > 2021) | ((my_data['YEAR'] == 2021) & (my_data['MONTH'] >= 2))]
after_coup.head()
```

```python
admin1_list = after_coup['ADMIN1'].unique()
print(admin1_list)

#replace Shan-North with Shan (North)
after_coup['ADMIN1'] = after_coup['ADMIN1'].replace({'Shan-North': 'Shan (North)'})
after_coup['ADMIN1'] = after_coup['ADMIN1'].replace({'Shan-South': 'Shan (South)'})
after_coup['ADMIN1'] = after_coup['ADMIN1'].replace({'Shan-East': 'Shan (East)'})

after_coup = my_data[(my_data['YEAR'] > 2021) | ((my_data['YEAR'] == 2021) & (my_data['MONTH'] >= 2))]
after_coup.head()

#group by ADMIN1 and sum events and fatalities
grouped_data = after_coup.groupby('ADMIN1').agg({'EVENTS': 'sum', 'FATALITIES': 'sum'}).reset_index()
grouped_data.head()

#replace Region (Other) with Region-Other through regex
grouped_data['ADMIN1'] = grouped_data['ADMIN1'].apply(lambda x: re.sub(r'(.*)-(.*)', r'\1 (\2)', x))

grouped_data.to_csv("myanmar_post_coup_events_fatalities_by_admin1.csv", index=False)
```
